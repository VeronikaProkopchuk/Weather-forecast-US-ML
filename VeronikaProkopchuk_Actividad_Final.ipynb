## Definición de funciones

```python
def get_city_frame(city_name,df_humidity,df_pressure,df_temperature,df_weather_description,df_wind_direct,df_wind_speed):
    '''
    Function that creates DataFrame of the selected city
    with data of humidity, pressure, temperature, weather description, wind direction and wind speed
    Input: 
       - city name in the format "City name";
       - 6 pandas dataframes to extract information
    Output: pandas dataframe
    '''
    city_df = (df_humidity[['datetime',city_name]].merge(df_pressure[['datetime',city_name]],on='datetime').merge(df_temperature[['datetime',city_name]], on='datetime').merge(df_weather_description[['datetime',city_name]], on='datetime').merge(df_wind_direct[['datetime',city_name]], on='datetime').merge(df_wind_speed[['datetime',city_name]], on='datetime'))
    city_df.columns=['Datetime','Humidity','Pressure','Temperature','Weather_description','Wind_direction','Wind_speed']

    return(city_df)
```

```python
def fill_nans(city_df):
    '''
    Function that fills nans
    Input: pandas dataframe
    Output: pandas dataframe
    '''
    for column_name in city_df.columns.values.tolist():
        city_df[column_name].fillna(method='bfill',inplace=True)
        city_df[column_name].fillna(method='ffill',inplace=True)
    
    return(city_df)
```

```python
def convert_to_celsius(city_df):
    '''
    Function that converts temperature from degree Fahrenheit to degree Celsius
    Input: pandas dataframe
    Output: pandas dataframe
    '''
    city_df['Temperature']=city_df['Temperature'].apply(lambda x:x - 273.15)

    return(city_df)
```

```python
def separate_time_columns(city_df):
    '''
    Function that creates separate columns for year,month,day and hour from the Datetime column
    Input: pandas dataframe
    Output: pandas dataframe
    '''
    city_df.Datetime = pd.to_datetime(city_df.Datetime)
    city_df['year']=pd.DatetimeIndex(city_df.Datetime).year
    city_df['month']=pd.DatetimeIndex(city_df.Datetime).month
    city_df['day']=pd.DatetimeIndex(city_df.Datetime).day
    city_df['hour']=pd.DatetimeIndex(city_df.Datetime).hour

    return(city_df)
```

## Eliminamos datos que no nos sirven

```python
def drop_columns(data, columns):
    '''
    Function that deletes the indicated columns (in a list) in the given dataframe
    Input:
        - data: pandas dataframe
        - columns: list
    Output: pandas dataframe
    '''
    data.drop(columns, axis=1, inplace=True)
    return
```

## Hacemos train test split

```python
from sklearn.model_selection import train_test_split
```

```python
y=city_new['Temperature']
X=city_new.drop(['Temperature'],axis=1)

validation_split = 0.30
seed = 250

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=validation_split,random_state=seed)
```

## Modelo Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
```

```python

dt_reg=DecisionTreeRegressor(random_state=250)
dt_model=GridSearchCV(dt_reg,param_grid={'max_depth':[10,11,12,13]})
```

```python
dt_model.fit(X_train,y_train)
```

```python
dt_model.best_estimator_
```

```python
dt_model.best_score_
```

```python
dt_reg=DecisionTreeRegressor(max_depth=13, random_state=250)
dt_reg.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
dt_y_predict = dt_reg.predict(X_test)
dt_mae = mean_absolute_error(y_test, dt_y_predict)
dt_mse = mean_squared_error(y_test, dt_y_predict)
print('MAE: ', dt_mae)
print('MSE: ', dt_mse) 
```

```python
dt_score = dt_reg.score(X_train,y_train)
print(dt_score)
```

## Importancia de las caracteristicas
